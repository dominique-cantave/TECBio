```python
import glob
import pandas as pd
import numpy as np
from sklearn.metrics import jaccard_similarity_score
import matplotlib.pyplot as plt
import os
import random


indexname = 'DJ'
year = 1925

stocks = pd.read_csv('stocklist.csv', sep=',', index_col=0)
stocks = stocks.iloc[:,0].values.tolist()
full_cols = ['in','age']
full_cols.extend(stocks)

datafiles = glob.glob("adjacency matrix/range_*.csv")
correlations = [None] * len(datafiles)

for f in datafiles:
    a = int(f.split(os.path.sep)[-1].split('_')[1][0:-4])
    temp_df = pd.read_csv(f, sep=',', index_col=0)
    df = pd.DataFrame(index=stocks, columns=stocks)
    for i in temp_df.index.values:
        for j in temp_df.columns.values:
            df.loc[i][j] = int(temp_df.at[i,j])
    
    df.fillna(0,inplace=True)
    correlations[a] = df
```

```python
def jaccard(df1, df2):
    dim1, dim2 = df1.shape
    tot = dim1*dim2
    k=0
    for i in range(dim1):
        for j in range(dim2):
            if df1.iat[i,j] == df2.iat[i,j]:
                k += 1
    
    return float(k)/tot
```

```python
corrtype = '0.3 pearson'
sims0 = []
sims100 = []
sims233 = []
deg = []

for i in range(len(correlations)):
    sim0 = jaccard2(correlations[i],correlations[0])
    sims0.append(sim0)
    sim100 = jaccard2(correlations[i],correlations[100])
    sims100.append(sim100)
    sim233 = jaccard2(correlations[i],correlations[233])
    sims233.append(sim233)
    
    
fig = plt.figure(1,figsize=(10,10))
fig.clf()
fig.add_subplot(3,1,1)
plt.plot(range(len(sims0)),sims0, 'k', color='#2B2ACC') 
plt.title('relative to t=0')
plt.xlabel('timestep')
plt.ylabel('pearson correlation')

fig.add_subplot(3,1,2)
plt.plot(range(len(sims100)),sims100, 'k', color='#2B2ACC') 
plt.title('relative to t=100')
plt.xlabel('timestep')
plt.ylabel('pearson correlation')

fig.add_subplot(3,1,3)
plt.plot(range(len(sims233)),sims233, 'k', color='#2B2ACC') 
plt.title('relative to t=end')
plt.xlabel('timestep')
plt.ylabel('pearson correlation')
plt.tight_layout()


plt.savefig('thresh={}-1.pdf'.format(corrtype))

# plt.plot(range(len(deg)), deg)
```

```python
simsrand = []
random.seed(5)

rand_df = pd.DataFrame(index=stocks,columns=stocks)
for i in stocks:
    for j in stocks:
        if i == j:
            rand_df.loc[i,j] = 1
        else:
            if random.random() < 0.2:
                rand_df.loc[i,j] = 1
            else:
                rand_df.loc[i,j] = 0
            
for i in range(len(correlations)):
    sim = jaccard_similarity_score(correlations[i],rand_df)
    simsrand.append(sim)
    
plt.plot(range(len(simsrand)),simsrand)
plt.savefig('thresh={}-rand1.pdf'.format(corrtype))
```

```python
simsprev = []
for i in range(1,len(correlations)):
    sim = jaccard_similarity_score(correlations[i],correlations[i-1])
    simsprev.append(sim)

plt.plot(range(len(simsprev)),simsprev)
```
